```python
def cng_date(x):
  import datetime
  a = x.split(':')
  day = a[0]
  hour = a[1]
  if hour == '24':
    hour = '00:00:00'
    final = pd.to_datetime(day+" "+hour) + datetime.timedelta(days =1)
    # 시간 더하고 빼기
  else:
    hour = hour+ ':00:00'
    final = pd.to_datetime(day+" "+hour)
  return final

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(cng_date)
```
