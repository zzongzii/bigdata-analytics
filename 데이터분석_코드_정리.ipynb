```python
df['newLimit'] = df.Credit_Limit.map(lambda x : 1 if x>=4500 else 0)
Ans = df['newLimit'].value_counts()
# 계속 map, lambda를 활용해서 뽑는거.
```

```python
def check(x):
    if x.Marital_Status =='Married' and x.Card_Category =='Platinum':
        return 1
    else:
        return 0


df['newState'] = df.apply(check,axis=1)

Ans  = df['newState'].value_counts()
# axis = 1, 이라는 것으로 가로를 기준으로 함수에 적용한다는 뜻, apply로 def한 함수 적용하는법
# 밑의 경우처럼 한 컬럼을 기준으로 apply 하는 경우는 괜찮지만 아니라면, apply 안에 axis = 1이라는 것을 꼭 넣어줘야 한다.
```

```python
# time_series
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
```

```python
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(":")[0]
    
    if hour =='24':
        hour ='00:00:00'
        
        FinalDate = pd.to_datetime(date +" "+hour) +datetime.timedelta(days=1)
        
    else:
        hour = hour +':00:00'
        FinalDate = pd.to_datetime(date +" "+hour)
    
    return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)

Ans = df

# 시간을 datetime으로 변경하기
```

```python
df['dayName']  =df['(년-월-일:시)'].dt.day_name()
df['dayName']
# 요일 컬럼 생성 
```

```python
Ans = pd.concat([df3,df4],join='inner')
# concat inner join 하는법
```

```python
pd.concat([df3,df4],join='outer').fillna(0)
# outer join, 뒤에 fillna 넣지 않으면 nan 처리됨. 
```

```python
pd.concat([df3,df4],join='outer',axis = 1).fillna(0)
# outer join, 뒤에 fillna 넣지 않으면 nan 처리됨. 
```
