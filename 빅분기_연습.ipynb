````python
import pandas  as pd
import numpy as np
a = 'https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv'
df = pd.read_csv(a,encoding = 'euc-kr')
````

````python
df.shape
````

````python
df.head()
````

````python
# 56~63
# apply, map
dic = {'Unknown' : 'N',
       'Less than $40K' : 'a',
       '$40K - $60K' : 'b',
       '$60K - $80K' : 'c',
       '$80K - $120K' : 'd',
       '$120K +': 'e'}
df['n_income'] = df.Income_Category.map(lambda x: dic[x])
df['n_income'].head()
````

````python
def cng_cat(x):
  if x == 'Less than $40K':
    return 'a'
  elif x == 'Unknown':
    return 'N'
  elif x == '$40K - $60K':
    return 'b'
  elif x == '$60K - $80K':
    return 'c'
  elif x == '$80K - $120K':
    return 'd'
  elif x == '$120K +':
    return 'e'

# df['n_income2'] = df.Income_Category.apply(lambda x : cng_cat(x))
# 위 아래 둘다 결과 같음
df['n_income2'] = df.Income_Category.apply(cng_cat)
df['n_income2']
````

````python
df['age_stage'] = df.Customer_Age.map(lambda x : str(x//10*10) + '대')
a = df.groupby('age_stage').size()
a
````

````python
df['new_edu'] = df.Education_Level.map(lambda x : 1 if 'Graduate' in x else 0)
df.new_edu.value_counts()
````

````python
df['n_limit'] = df.Credit_Limit.map(lambda x : 1 if x >= 4500 else 0)
df.n_limit.value_counts
````

````python
def one_ze(x):
  if x.Marital_Status =='Married' and x.Card_Category == 'Platinum':
    return 1
  else :
    return 0

df['n_state'] = df.apply(one_ze,axis = 1) # apply 안에 함수로 넣기., 가로로 axis = 1 로 넣기
df.n_state.value_counts()
````

````python
df.loc[df['Gender']=='M','Gender'] = 'male'
df.loc[df['Gender']=='F','Gender'] = 'female'
````

````python
df.Gender.value_counts()
````

````python
def cng_gen(x):
  if x == 'M':
    return 'male'
  elif x == 'F':
    return 'female'

df['Gender']=df.Gender.apply(cng_gen)
df.Gender.value_counts()
````
